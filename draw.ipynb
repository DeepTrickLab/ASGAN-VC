```python
import random
import numpy as np
from matplotlib import pyplot as plt
from math import exp
```

```python
def generate_pic(model_name,similarity,ticks,title="pic"):
    color = [(random.randint(0,10)/10,random.randint(0,10)/10,random.randint(0,10)/10) for x in range(len(similarity))]
    plt.bar(
        model_name,
        similarity, 
        width=0.5, 
        bottom=None, 
        align='center', 
        color=color
    )
    plt.xticks(rotation='vertical')
    plt.ylim(ticks[0],ticks[-1])
    plt.yticks(ticks)
    plt.axhline(y=similarity[0], c="r", ls="--", lw=2)
    plt.axhline(y=min(similarity), c="g", ls="--", lw=2)
    plt.title(title,pad=20)
    plt.show() 
```

```python
import matplotlib
import random
import matplotlib.pyplot as plt

#

data = [98.03,89.77,92.99,93.43,93.39,92.96,93.95,94.18,94.24,93.05,94.41,95.08,94.2,75.87,87.27]
model_name = ['Real Data', 'AutoVC3 SNGAN', 'AutoVC3 GAN', 'AutoVC2 BiGAN', 'AutoVC3 BIGAN',
              'AutoVC2 SNGAN', 'AutoVC2 GAN', 'AutoVC SNGAN', 'AutoVC BIGAN',
              'AutoVC3', 'AutoVC GAN', 'AutoVC2', 'AutoVC',
              'VQVC+','AgainVC'
             ]
model_name.reverse()
data.sort(reverse=True)
data.reverse()
#
fig, ax = plt.subplots()
b = ax.barh(range(len(model_name)), data, color='#6699CC')

for rect in b:
    w = rect.get_width()
    ax.text(w, rect.get_y()+rect.get_height()/2, '%.2f' %float(w), ha='left', va='center')

ax.set_yticks(range(len(model_name)))
ax.set_yticklabels(model_name)
plt.xticks(())
plt.title('AUC', loc='center', fontsize='25',fontweight='bold', color='red')
plt.show()
```
